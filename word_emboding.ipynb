```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
vocab_size = 10000
```

```python
one_hot_reps = [one_hot(words,vocab_size)for words in sent]
one_hot_reps
```

```python
## word embodding
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_len = 8
embodded_docs = pad_sequences(one_hot_reps,padding="pre",maxlen=sent_len)
embodded_docs
```

```python
#feature rpre
dim = 10
model = Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embodded_docs)
```
